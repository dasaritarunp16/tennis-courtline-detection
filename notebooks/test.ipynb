```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../images/img1.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Blur = cv2.GaussianBlur(img_gray, (5,5), 0)
plt.imshow(Blur, cmap='gray')
plt.show()
edges = cv2.Canny(Blur, 100, 150)
plt.imshow(edges, cmap = 'gray')
plt.show()


houghlinesdetected = cv2.HoughLinesP(edges, 1, np.pi/180, threshold = 100, minLineLength = 100, maxLineGap = 10)
for line in houghlinesdetected:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1,y1), (x2,y2), (0,255,0), 3)

cv2.imwrite('houghlines.png', img)



```
